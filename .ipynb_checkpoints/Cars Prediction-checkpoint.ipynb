```python
#step1: import module------>LabelEncoder
from sklearn.preprocessing import LabelEncoder
#step2: initialize a class
le=LabelEncoder()
#step3: apply encoding
le.fit_transform(df_cat["fuel-type"])
```

```python
le.classes_
```

```python
df_cat
```

```python
#apply label encoding on all cols
```

```python
le=LabelEncoder()
for i in df_cat:
    df_cat[i]=le.fit_transform(df_cat[i])
```
